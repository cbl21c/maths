```python
import math
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

# if using a Jupyter notebook, include:
%matplotlib inline
#plt.style.use('seaborn-whitegrid')

# function to return value that is passed as argument
def f(x):
    return x

```

```python
a = widgets.IntSlider(value=20, min=20, max=200)
widgets.interactive_output(f, {'a': a})
display(a)

```

```python
NUM_TRIALS = a.value

# pre-compute (1/2)^20
p = (1/2)**NUM_TRIALS

# build the distribution
P = [0] * (NUM_TRIALS + 1)
for k in range(NUM_TRIALS + 1):
    P[k] = math.comb(NUM_TRIALS, k) * p
    
krange = np.linspace(0, NUM_TRIALS, NUM_TRIALS + 1)
# choose a range of 100 points around the centre to plot
# if fewer than 100, plot the entire distribution
xmin = int(NUM_TRIALS / 2 - 50)
xmax = xmin + 100
if NUM_TRIALS < 100:
    xmin = 0
    xmax = NUM_TRIALS

# plot the distribution
fig = plt.figure()
ax = plt.axes()
ax.plot(krange[xmin:xmax], P[xmin:xmax])

```
